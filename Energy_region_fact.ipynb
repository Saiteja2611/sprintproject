```python
from pyspark.sql.functions import sum as _sum

df_telemetry = spark.table("silver_smart_meter_telemetry")

df_fact_energy = (
    df_telemetry
    .groupBy("region", "city")
    .agg(_sum("kWh").alias("TotalConsumption"))
)

df_fact_energy.write \
    .mode("overwrite") \
    .format("delta") \
    .save("abfss://sedpcontainer@sedpstorageaccount.dfs.core.windows.net/Gold/gold_fact_energy_consumption_delta")

spark.sql("""
    CREATE TABLE IF NOT EXISTS gold_fact_energy_consumption
    USING DELTA
    LOCATION 'abfss://sedpcontainer@sedpstorageaccount.dfs.core.windows.net/Gold/gold_fact_energy_consumption_delta'
""")
```
